Monthly

```python
p1 = "https://freemeteo.co.uk/weather/seoul/history/monthly-history/?gid=1835848&station=10565&month="
month_range = range(1, 13)
p2 = "&year="
year_range = range(2012, 2023)
p3 = "&language=english&country=south-korea"
```

```python
path = p1 + str(1) + p2 + str(2012) + p3
pd.read_html(path)[6]
```

```python
for year in year_range:
    for month in month_range:
        path = p1 + str(month) + p2 + str(year) + p3
        
        pd.read_html(path)[6].to_csv('monthly/'+str(year)+'_'+str(month)+'.csv', index=False)
```

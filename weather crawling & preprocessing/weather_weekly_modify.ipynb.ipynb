## Daily

```python
import pandas as pd
daily_df = pd.read_csv('daily_save1.csv')
daily_df
```

```python
import datetime
print(datetime.date(2015, 12, 19) - datetime.date(2012, 1, 1) + datetime.timedelta(days=1))
print(datetime.date(2022, 4, 30) - datetime.date(2015, 12, 27) + datetime.timedelta(days=1))
```

```python
# week 정의
import datetime
start_day1 = datetime.date(2012, 1, 1)
year1 = 2012
week1 = 1

week_list1 = list()

i = 0
for i in range(1449):
    date1 = start_day1 + datetime.timedelta(days=i)
    
    date = {'date':[date1],
            'Year':[year1],
            'week':[week1]}
    
    week_df1 = pd.DataFrame(date)
    week_list1.append(week_df1[:])
    
    if i % 7 == 6:
        week1 += 1
        if week1 == 53:
            year1 += 1
            week1 = 1
week_all1 = pd.concat(week_list1)

### 구분
start_day2 = datetime.date(2015, 12, 27)
year2 = 2015
week2 = 52

week_list2 = list()

i = 0
for i in range(2317):
    date2 = start_day2 + datetime.timedelta(days=i)
    
    date = {'date':[date2],
            'Year':[year2],
            'week':[week2]}
    
    week_df2 = pd.DataFrame(date)
    week_list2.append(week_df2[:])
    
    if i % 7 == 6:
        week2 += 1
        if week2 == 53:
            year2 += 1
            week2 = 1
week_all2 = pd.concat(week_list2)

week_all = pd.concat([week_all1, week_all2])
week_all
```

```python
# date column 생성
daily_df['date'] = 0
for i in range(len(daily_df['date'])):
    daily_df.loc[i, 'date'] = datetime.date(int(daily_df.loc[i, 'Year']), 
                                            int(daily_df.loc[i, 'month']),
                                            int(daily_df.loc[i, 'day']))
daily_df
```

```python
interested = ['Temperature', 'Relative Temperature', 'Wind', 'Rel. humidity', 'Dew Point', 'Pressure', 'date']
```

```python
week_daily_df = pd.merge(daily_df[interested], week_all, on = ['date'])
week_daily_df
```

```python
interested = ['Year', 'week', 'Temperature', 'Relative Temperature', 'Wind', 'Rel. humidity', 'Dew Point', 'Pressure']
week_daily_df[interested]
```

```python
week_daily_df[interested].to_csv('weekly_daily.csv', index=False)
```

#### 가공 필요: R에서 진행하기
- 불쾌지수
- 바람냉각지수
- 보퍼트 풍력 지수
- 체감온도
- 일교차

````python
import pandas as pd

Year_range = range(2012, 2023)
month_range = range(1, 13)


monthly_list = list()

for Year in Year_range:
    for month in month_range:
        if Year == 2022 and month == 5:
            break

        date = str(Year) + '_' + str(month)

        df2 = pd.read_csv('monthly/'+date+'.csv')
        df2['Year'] = df2['Date'].apply(lambda x : x.split('/')[2])
        df2['month'] = df2['Date'].apply(lambda x : x.split('/')[0])
        df2['day'] = df2['Date'].apply(lambda x : x.split('/')[1])
        df2['Description'] = df2['Description'].apply(lambda x : '-' if x=='weather events not reported' else x)

        interested = ['Year', 'month', 'day', 'Daily minimum temperature', 'Daily maximum temperature', 'Description']
        monthly_list.append(df2[interested])

monthly_weather = pd.concat(monthly_list)
monthly_weather
````

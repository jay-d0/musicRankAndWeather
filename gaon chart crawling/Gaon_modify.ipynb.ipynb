```python
import pandas as pd
from glob import glob

gaon_path_ = 'gaon_chart'
gaon_path = glob(gaon_path_ + "/*.csv")

gaon_list = list()

for path in gaon_path:
    
    df = pd.read_csv(path)
    df['date'] = path
    df['Year'] = df['date'].apply(lambda x : x.split('_')[2])
    df['week'] = df['date'].apply(lambda x : x.split('_')[3].split('w')[0])
    interested = ['Year', 'week', 'Ranking.1', 'Title / Artist.1', 'Production']
    gaon_list.append(df[interested])    

all_chart = pd.concat(gaon_list)

all_chart
```

```python
all_chart['Year'] = all_chart['Year'].apply(lambda x : int(x))
all_chart['week'] = all_chart['week'].apply(lambda x : int(x))
all_chart = all_chart.sort_values(by = ['Year', 'week'], ascending=True)
```

```python
all_chart
```

```python
import pandas as pd
from glob import glob

rank_path_ = 'gaon_rank'
rank_path = glob(rank_path_ + "/*.csv")

rank_list = list()

for path in rank_path:
    
    df = pd.read_csv(path)
    df['date'] = path
    df['Year1'] = df['date'].apply(lambda x : x.split('\\')[1].split('_')[0])
    df['week1'] = df['date'].apply(lambda x : x.split('\\')[1].split('_')[1].split('.')[0])
    rank_list.append(df)

all_rank = pd.concat(rank_list)

all_rank
```

```python
all_rank['Year1'] = all_rank['Year1'].apply(lambda x : int(x))
all_rank['week1'] = all_rank['week1'].apply(lambda x : int(x))
all_rank = all_rank[['Year1', 'week1', 'rank_change']].sort_values(by=['Year1', 'week1'], ascending=True)
```

```python
all_rank
```

```python
df = pd.concat([all_chart, all_rank], axis=1)
```

```python
df[['Year', 'week', 'Ranking.1', 'rank_change', 'Title / Artist.1', 'Production']].to_csv('gaon_chart.csv', index=False)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

path_s = "http://www.gaonchart.co.kr/main/section/chart/online.gaon?nationGbn=T&serviceGbn=ALL&targetTime="
path_m = "&hitYear="
path_e = "&termGbn=week"
```

```python
tT_list = range(1, 53)
Year_list = range(2012, 2023)
```

```python
pd.read_html(path_s + str(17) + path_m + str(2022) + path_e)[0]
```

```python
import pandas as pd
import time

for Year in Year_list:
    for tT in tT_list:
        df = pd.read_html(path_s + str(tT) + path_m + str(Year) + path_e)[0]
        df.to_csv('gaon_chart/chart_'+str(Year)+'_'+str(tT)+'week.csv', index=False)
        time.sleep(1)
        if Year == 2022 and tT == 19:
            break
```

```python
for Year in Year_list:
    for tT in tT_list:
        res = requests.get(path_s + str(tT) + path_m + str(Year) + path_e)
        soup = BeautifulSoup(res.content, 'html.parser')

        changes = soup.find_all('td', 'change')

        rank_change = list()
        for change in changes:
            if str(change.find('span')) != '<span>-</span>':
                rank_change.append(str(change.find('span')).split('\"')[1])
            else:
                rank_change.append('-')
        
        df = pd.DataFrame(rank_change)
        df.columns = ['rank_change']
        df.to_csv('gaon_rank/'+str(Year)+'_'+str(tT)+'.csv', index=False)
```

```python
df_list = list()
for Year in Year_list:
    for tT in tT_list:
        res = requests.get(path_s + str(tT) + path_m + str(Year) + path_e)
        soup = BeautifulSoup(res.content, 'html.parser')

        singers = soup.find_all('p', 'singer')

        singer_list = list()
        for singer in singers:
            singer_list.append(singer.get_text())

        week_df = pd.DataFrame(singer_list)
        week_df.columns = ['singer']
        
        df_list.append(week_df)
        if Year == 2022 and tT == 18:
            break
        
singers_list = pd.concat(df_list)
```
